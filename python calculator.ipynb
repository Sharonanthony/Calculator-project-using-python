```python
#Python Calculator

#importing logging
import logging 

logging.basicConfig(filename='trytest.log', level=logging.DEBUG, format='%(asctime)s-%(levelname)s- %(message)s')

#importing reduce from functool
from functools import reduce

class calculator:
    def addition(self, *args):
        logging.info("Addition is being performed")
        return sum(args)

    def subtraction(self, *args):
        logging.info("Subtraction is being performed")
        return reduce(lambda a, b: a - b, args)

    def multiplication(self, *args):
        logging.info("Multiplication is being performed")
        return reduce(lambda a, b: a * b, args)

    def division(self, *args):
        logging.info("Division is being performed")
        if 0 in args[1:]:
            print("Error! Division by zero is not possible!")
        return reduce(lambda a, b: a / b, args)

    def power(self, a, n):
        logging.info("Power is being found based on the user input")
        if a == 0:
            print("Error! Enter a non-zero number!")
            return None
        if n == 0:
            return 1
        else:
            return a * self.power(a, n - 1)

    def fibonacci(self, n):
        logging.info("Fibonacci is being found based on the user input")
        if n < 0:
            raise ValueError("Enter a non-negative number")
        if n == 1 or n == 0:
            return n
        else:
            return self.fibonacci(n - 1) + self.fibonacci(n - 2)

    def factorial(self, n):
        logging.info("Factorial is being found based on the user input")
        if n < 0:
            raise ValueError("Enter a non-negative number")
        if n == 1 or n == 0:
            return 1
        else:
            return n * self.factorial(n - 1)

logging.info("This is the start of the program")
print("Choose the operation you want to do!")
print("1. Addition")
print("2. Subtraction")
print("3. Multiplication")
print("4. Division")
print("5. Power")
print("6. Fibonacci series")
print("7. Factorial")

while True:
    logging.info("Asking the user to enter their choice")
    choice = input("Enter your choice (1/2/3/4/5/6/7): ")

    if choice in ("1", "2", "3", "4"):
        logging.info("Trying to check whether the choice is 1/2/3/4")
        try:
            numbers = [float(num) for num in input("Enter the numbers separated by spaces: ").split()]
        except ValueError:
            print("Invalid input. Please enter numbers separated by spaces.")
            continue

        first_obj = calculator()

        if choice == "1":
            logging.info("The user entered choice 1")
            print("The result of the addition of the given numbers is:", first_obj.addition(*numbers))
        elif choice == "2":
            logging.info("The user entered choice 2")
            print("The result of the subtraction of the given numbers is:", first_obj.subtraction(*numbers))
        elif choice == "3":
            logging.info("The user entered choice 3")
            print("The result of the multiplication of the given numbers is:", first_obj.multiplication(*numbers))
        elif choice == "4":
            logging.info("The user entered choice 4")
            try:
                print("The result of the division of the given numbers is:", first_obj.division(*numbers))
            except ZeroDivisionError as e:
                print(e)

    elif choice=="5":
        logging.info("The user entered choice 5")
        try:
            a = float(input("Enter the coefficient: "))
            n = int(input("Enter the exponent: "))
        except ValueError:
            print("Invalid input. Please enter numbers.")
            continue

        first_obj = calculator()
        result = first_obj.power(a, n)
        if result is not None:
            print("The power of the given number is:", result)

    elif choice=="6":
        logging.info("The user entered choice  6")
        try:
            n = int(input("Enter a number: "))
        except ValueError:
            print("Invalid input. Please enter a number.")
            continue

        first_obj = calculator()
        print("The Fibonacci of the given number is:", first_obj.fibonacci(n))
    
    elif choice=="7":
        logging.info("The user entered choice 7")
        try:
            n = int(input("Enter a number: "))
        except ValueError:
            print("Invalid input. Please enter a number.")
            continue

        first_obj = calculator()
        print("The factorial of the given number is:", first_obj.factorial(n))
    
    else:
        print("Invalid Input")
    logging.info("Trying to check whether the user wants to perform more operations")
    try:
        next_calculation = input("Let's do next calculation? (yes/no): ")
        if next_calculation.lower() == "no":
            break
        else:
            raise ValueError("Enter a valid answer!")
    except ValueError as e:
        print(e)
```
